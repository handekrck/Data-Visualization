```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
from matplotlib.pyplot import figure
from sklearn.preprocessing import MinMaxScaler
import warnings
```

```python
data = pd.read_excel("timeseries.xlsx", index_col= 0)
data = data.dropna()
data.head()
```

```python
data_t = data.T
data_t.head()
```

```python
fransa = data.iloc[6]
ingiltere = data.iloc[20]
italya = data.iloc[8]
norveç = data.iloc[18]
```

```python
grup = pd.DataFrame([fransa,ingiltere,italya,norveç]) 
grupT = grup.T
```

```python
figure(figsize=(12, 10), dpi=100)

plt.subplot(2,2,1)
plt.plot(data_t,color='grey')
plt.plot(fransa, color="red")
plt.title("fransa")
plt.xticks(rotation=90)

plt.subplot(2,2,2)
plt.plot(data_t,color='grey')
plt.plot(ingiltere, color="orange")
plt.title("ingiltere")
plt.xticks(rotation=90)

plt.subplot(2,2,3)
plt.plot(data_t,color='grey')
plt.plot(italya, color="yellow")
plt.title("italya")
plt.xticks(rotation=90)

plt.subplot(2,2,4)
plt.plot(data_t,color='grey')
plt.plot(norveç, color="cyan")
plt.title("norveç")
plt.xticks(rotation=90)
```
